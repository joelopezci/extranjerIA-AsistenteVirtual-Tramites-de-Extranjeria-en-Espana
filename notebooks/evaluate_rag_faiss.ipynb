```python
import os
import pandas as pd
from IPython.display import display
import ipywidgets as widgets
import sys

# Ruta del proyecto
PROJECT_ROOT = os.path.abspath(os.path.join(os.getcwd(), ".."))
sys.path.append(PROJECT_ROOT)

# Verificar ruta
print("Ruta del proyecto:", PROJECT_ROOT)
```

```python
from src.evaluate_rag_faiss import evaluate_rag_performance
```

```python
# Definir la ruta del archivo de evaluaci√≥n
input_excel_path = os.path.join(PROJECT_ROOT, "data", "corpus_extranjerIA_muestra.xlsx")

# Leer el archivo de Excel para inspecci√≥n
df = pd.read_excel(input_excel_path)
display(df.head())
```

```python
# Definir listas de opciones disponibles
embedding_models = [
    "distiluse-base-multilingual-cased-v1",
    "mrm8488/multilingual-e5-large-ft-sts-spanish-matryoshka-768-16-5e"
]
embedding_providers = ["huggingface"]

llm_providers = ["groq", "huggingface"]
llm_models = ["llama-3.2-1b-preview", "llama-3.2-3b-preview", "mixtral-8x7b-32768"]
# llm_models = ["llama-3.2-1b-preview", "LenguajeNaturalAI/leniachat-gemma-2b-v0", "llama-3.2-3b-preview", "deepseek-ai/DeepSeek-R1-Distill-Qwen-7B", "mixtral-8x7b-32768"]

# Crear widgets para la selecci√≥n de par√°metros
embedding_model_dropdown = widgets.Dropdown(
    options=embedding_models,
    description="Emb. Model:",
    style={'description_width': 'initial'}
)

embedding_provider_dropdown = widgets.Dropdown(
    options=embedding_providers,
    description="Emb. Provider:",
    style={'description_width': 'initial'}
)

llm_provider_dropdown = widgets.Dropdown(
    options=llm_providers,
    description="LLM Provider:",
    style={'description_width': 'initial'}
)

llm_model_dropdown = widgets.Dropdown(
    options=llm_models,
    description="LLM Model:",
    style={'description_width': 'initial'}
)

# Bot√≥n para iniciar la evaluaci√≥n
start_evaluation_button = widgets.Button(
    description="Iniciar Evaluaci√≥n",
    button_style="success"
)
```

```python
def run_evaluation(button):
    # Obtener valores seleccionados de los widgets
    embedding_model = embedding_model_dropdown.value
    embedding_provider = embedding_provider_dropdown.value
    llm_provider = llm_provider_dropdown.value
    llm_model = llm_model_dropdown.value

    # Reemplazar caracteres problem√°ticos para nombres de archivos
    safe_embedding_model = embedding_model.replace("/", "_")
    safe_llm_model = llm_model.replace("/", "_")

    # Crear un nombre de archivo basado en la configuraci√≥n
    output_excel_name = f"evaluation_faiss_{safe_embedding_model}_{embedding_provider}_{safe_llm_model}_{llm_provider}.xlsx"
    output_excel_path = os.path.join(PROJECT_ROOT, "data", "evaluations", output_excel_name)

    # Crear el directorio si no existe
    os.makedirs(os.path.dirname(output_excel_path), exist_ok=True)

    print(f"Evaluaci√≥n iniciada con:")
    print(f"- Modelo de embeddings: {embedding_model}")
    print(f"- Proveedor de embeddings: {embedding_provider}")
    print(f"- LLM usado: {llm_model}")
    print(f"- Proveedor de LLM: {llm_provider}")
    print(f"Guardando resultados en: {output_excel_path}")

    # Ejecutar la evaluaci√≥n RAG
    try:
        evaluate_rag_performance(
            input_excel=input_excel_path,
            output_excel=output_excel_path,
            embedding_model=embedding_model,
            embedding_provider=embedding_provider,
            llm_provider=llm_provider,
            llm_model=llm_model
        )
        print("\n‚úÖ Evaluaci√≥n completada exitosamente.")
    except Exception as e:
        print(f"‚ùå Error durante la evaluaci√≥n: {e}")

# Asociar la funci√≥n al bot√≥n despu√©s de definirla
start_evaluation_button.on_click(run_evaluation)
```

```python
# Funci√≥n para mostrar la interfaz
def display_interface():
    display(
        widgets.VBox([
            widgets.HBox([embedding_model_dropdown, embedding_provider_dropdown]),
            widgets.HBox([llm_provider_dropdown, llm_model_dropdown]),
            start_evaluation_button
        ])
    )

# Mostrar la interfaz de selecci√≥n
display_interface()
```

```python
# Directorio donde se guardan los archivos de resultados
RESULTS_DIR = os.path.join(PROJECT_ROOT, "data", "evaluations")

def show_results():
    """
    Muestra los resultados de evaluaci√≥n permitiendo seleccionar un archivo de resultados.
    """
    # Obtener lista de archivos disponibles en el directorio de resultados
    result_files = [f for f in os.listdir(RESULTS_DIR) if f.endswith('.xlsx')]

    if not result_files:
        print("No hay archivos de resultados disponibles.")
        return
    
    # Widget de selecci√≥n de archivo
    file_selector = widgets.Dropdown(
        options=result_files,
        description='Archivo:',
        style={'description_width': 'initial'},
        layout=widgets.Layout(width='50%')
    )

    def on_file_selected(change):
        """
        Carga y muestra el contenido del archivo seleccionado.
        """
        selected_file = os.path.join(RESULTS_DIR, file_selector.value)
        df = pd.read_excel(selected_file)
        print(f"\nMostrando resultados del archivo: {file_selector.value}")
        
        # Mostrar las primeras filas del dataframe
        display(df.head(10))

    # Bot√≥n para cargar el archivo seleccionado
    load_button = widgets.Button(
        description="Cargar resultados",
        button_style='success',
        icon='check'
    )
    
    load_button.on_click(on_file_selected)
    
    # Mostrar widgets
    display(file_selector, load_button)

# Llamar a la funci√≥n para mostrar la interfaz de selecci√≥n
show_results()
```
